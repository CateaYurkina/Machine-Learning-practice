```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Параметры распределения
mu = 0      # среднее
sigma = 1   # стандартное отклонение

# Генерация случайных чисел из нормального распределения
sample_size = 1000
data = np.random.normal(mu, sigma, sample_size)

# Построение гистограммы
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Гистограмма')

# Построение теоретической плотности вероятности
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)
pdf = norm.pdf(x, mu, sigma)
plt.plot(x, pdf, 'k', linewidth=2, label='Теоретическая плотность')

plt.title('Нормальное распределение')
plt.xlabel('Значение')
plt.ylabel('Плотность вероятности')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Параметры равномерного распределения
a = 0    # нижняя граница
b = 1    # верхняя граница

# Генерация случайных чисел из равномерного распределения
sample_size = 1000
data = np.random.uniform(a, b, sample_size)

# Построение гистограммы
plt.hist(data, bins=30, density=True, alpha=0.6, color='b', label='Гистограмма')

# Построение теоретической плотности вероятности
x = np.linspace(a, b, 100)
pdf = np.ones_like(x) / (b - a)  # плотность равномерного распределения
plt.plot(x, pdf, 'r', linewidth=2, label='Теоретическая плотность')

plt.title('Равномерное распределение')
plt.xlabel('Значение')
plt.ylabel('Плотность вероятности')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Вероятность успеха
p = 0.6

# Количество экспериментов
n = 1000

# Генерация данных: 0 или 1
data = np.random.binomial(1, p, n)

# Подсчет количества успехов и неудач
successes = np.sum(data)
failures = n - successes

# Построение столбчатой диаграммы
plt.bar(['Неудача (0)', 'Успех (1)'], [failures/n, successes/n], color=['red', 'green'])
plt.ylabel('Вероятность')
plt.title('Распределение Бернулли')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Параметры
n = 20      # число испытаний
p = 0.5     # вероятность успеха в каждом испытании

# Значения k (число успехов)
k = np.arange(0, n+1)

# Вероятности для каждого k
probabilities = binom.pmf(k, n, p)

# Визуализация
plt.bar(k, probabilities, color='skyblue')
plt.xlabel('Число успехов')
plt.ylabel('Вероятность')
plt.title(f'Биномиальное распределение (n={n}, p={p})')
plt.grid(True)
plt.show()
```

```python
import random

def linear_regression_train(X, y, learning_rate=0.01, epochs=1000):
    """
    Обучает линейную регрессию с помощью градиентного спуска.

    Parameters:
        X (list of list of float): матрица признаков (размер m x n)
        y (list of float): вектор целевых значений (размер m)
        learning_rate (float): скорость обучения
        epochs (int): число итераций

    Returns:
        weights (list of float): обученные веса модели
    """
    m = len(X)          # число примеров
    n = len(X[0])       # число признаков

    # Инициализация весов случайными значениями
    weights = [random.uniform(-0.5, 0.5) for _ in range(n + 1)]  # +1 для свободного члена

    for epoch in range(epochs):
        total_error = 0
        gradients = [0.0] * (n + 1)

        for i in range(m):
            # Добавляем свободный член к признакам
            x_i = [1.0] + X[i]
            y_pred = sum(w * x for w, x in zip(weights, x_i))
            error = y_pred - y[i]
            total_error += error ** 2

            # Обновляем градиенты
            for j in range(n + 1):
                gradients[j] += error * x_i[j]

        # Средние градиенты по всему набору данных
        for j in range(n + 1):
            gradients[j] /= m

        # Обновление весов
        for j in range(n + 1):
            weights[j] -= learning_rate * gradients[j]

        # Можно добавить вывод ошибки для мониторинга
        if epoch % 100 == 0:
            print(f"Epoch {epoch}: MSE={total_error/m:.4f}")

    return weights

# Пример использования:
# X = [[признак1, признак2, ...], ...]
# y = [значение1, значение2, ...]
# weights = linear_regression_train(X, y)
```

```python
import math
import random

def sigmoid(z):
    return 1 / (1 + math.exp(-z))

def logistic_regression_train(X, y, learning_rate=0.1, epochs=1000):
    """
    Обучает логистическую регрессию с помощью градиентного спуска.

    Parameters:
        X (list of list of float): матрица признаков (m x n)
        y (list of int): метки классов (0 или 1)
        learning_rate (float): скорость обучения
        epochs (int): число итераций

    Returns:
        weights (list of float): обученные веса модели
    """
    m = len(X)
    n = len(X[0])

    # Инициализация весов случайными значениями
    weights = [random.uniform(-0.5, 0.5) for _ in range(n + 1)]  # +1 для свободного члена

    for epoch in range(epochs):
        total_loss = 0
        gradients = [0.0] * (n + 1)

        for i in range(m):
            x_i = [1.0] + X[i]
            z = sum(w * x for w, x in zip(weights, x_i))
            pred = sigmoid(z)
            error = pred - y[i]
            total_loss += - (y[i] * math.log(pred + 1e-15) + (1 - y[i]) * math.log(1 - pred + 1e-15))

            # Обновление градиентов
            for j in range(n + 1):
                gradients[j] += error * x_i[j]

        # Средние градиенты
        for j in range(n + 1):
            gradients[j] /= m

        # Обновление весов
        for j in range(n + 1):
            weights[j] -= learning_rate * gradients[j]

        # Можно выводить ошибку каждые 100 эпох
        if epoch % 100 == 0:
            print(f"Epoch {epoch}: Log Loss={total_loss/m:.4f}")

    return weights

# Пример использования:
# X = [[признак1, признак2], ...]
# y = [0, 1, ...]
# weights = logistic_regression_train(X, y)
```
